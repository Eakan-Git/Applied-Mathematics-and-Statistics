### Khai báo thư viện

```python
from __future__ import division, print_function, unicode_literals
import numpy as np 
import matplotlib.pyplot as plt
```

# Ví dụ 1:

### Khai báo, biểu diễn dữ liệu trên đồ thị

```python
X = np.array([[0, 0.5, 1, 1.5, 2, 2.5]]).T
y = np.array([[0, 1.5, 3, 4.5, 6, 7.5]]).T
# Visualize data 
plt.plot(X, y, 'ro')
plt.axis([0, 3, 0, 10])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Nghiệm theo công thức

```python
one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis = 1)

A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)
print('w = ', w)

# Preparing the fitting line 
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(0, 7.5, 2)
y0 = w_0 + w_1*x0

# Drawing the fitting line 
plt.plot(X.T, y.T, 'ro')     # data 
plt.plot(x0, y0)               # the fitting line
plt.axis([0, 3, 0, 10])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Dự đoán Y từ X

```python
y1 = w_1*1.2 + w_0
y2 = w_1*2.6 + w_0

print(u'Nếu x = 1.2 thì có khả năng y = %.2f' %(y1))
print(u'Nếu x = 2.6 thì có khả năng y = %.2f' %(y2))
```
